```python
#Reading and displaying an image
import cv2
image = cv2.imread("paris.jpg", 1)
```

Show an image in a new window : (rgb colour variation)

```python
cv2.imshow('Original Image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Display an image : (generally used to show more images in a single frame)(BGR colour variation)

```python
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("paris.jpg")
plt.imshow(img)
plt.show()
```

Showing Shape and size of an image :

```python
print(img.size)
print(np.size(image))
print(img.shape)
print(np.shape(img))
print(img[0,0])
```

RGB to Gray Colour conversion : method 1

```python
gray = cv2.imread("paris.jpg", 0)
cv2.imshow("Grayscale Image" ,gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

RGB to Gray Colour Conversion : method 2

```python
img  = cv2.imread("paris.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale Image" ,gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Grayscale to Binary :

```python
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

titles = ['Original Image', 'Binary', 'Binary Inverse', 'Truncated', 'ToZero', 'ToZero Inverse']
images =[img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])

plt.xticks([]),plt.yticks([])
plt.show()
```
