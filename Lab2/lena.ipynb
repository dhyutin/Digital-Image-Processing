```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Lena Image
lena = cv2.imread("Lena.png", 1)
# converting BGR to RGB
image_rgb = cv2.cvtColor(lena, cv2.COLOR_BGR2RGB)

# Conversion of Lena to grayscale
lena_gray = cv2.imread("Lena.png", 0)
```

```python
# Resized lena images
img1 = cv2.resize(lena_gray, (128, 128))
img2 = cv2.resize(lena_gray, (64, 64))
img3 = cv2.resize(lena_gray, (32, 32))
img4 = cv2.resize(lena_gray, (16, 16))
```

```python
# Display images with different sizes
titles = ['Original','128X128', '64X64','32X32', '16X16']
images =[image_rgb, img1, img2, img3, img4]

plt.figure(figsize=(15,15))

for i in range(5):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])

plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# Display images with different intensity ranges
img_255 = lena_gray
img_127 = lena_gray // 2
img_63 = lena_gray // 4
img_31 = lena_gray // 8
img_15 = lena_gray // 16
img_7 = lena_gray // 32
img_3 = lena_gray // 64
img_1 = lena_gray // 128
```

```python
# Display images with intensity ranges
titles = ['0-255', '0-127', '0-63', '0-31', '0-15', '0-7','0-3', '0-1']
images =[img_255, img_127, img_63, img_31, img_15, img_7, img_3, img_1]

plt.figure(figsize=(15,15))

for i in range(8):
    plt.subplot(4,2, i+1)
    plt.imshow(images[i], cmap = 'gray')
    plt.title(titles[i])

plt.xticks([]),plt.yticks([])
plt.show()
```
