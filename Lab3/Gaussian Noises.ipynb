```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import math

# Function to calculate avg of images
def img_avg(img_list):
    sum = 0
    for i in img_list:
        sum = sum + i
    avg = sum / len(img_list)
    return avg

# Lena Image
lena = cv2.imread("Lena.png", 1)
# converting BGR to RGB
#image_rgb = cv2.cvtColor(lena, cv2.COLOR_BGR2RGB)

# Conversion of Lena to grayscale
gray = cv2.cvtColor(lena, cv2.COLOR_BGR2GRAY)
```

```python
# # finding Mean
# temp = lena
# temp = temp.flatten()
# avg = 0
# for i in range(len(temp)):
#     avg = avg + temp[i]
# mean = avg / len(temp)

# #finding standard deviation w.r.t mean
# def find_sd(mean):
#     variance = 0
#     for i in range(len(temp)):
#         diff = temp[i] - mean
#         sq = diff * diff
#         div = sq / len(temp)
#         variance = variance + div
#     sd = math.sqrt(variance)
#     return sd

# Adding Gaussian Noises
noise = []

for i in range(10):
    mean1 = np.random.randint(20)
    #sd = find_sd(mean1)
    sd = np.random.randint(50)
    #generate a random gaussian distribution with respective mean and standard deviation
    x = np.random.normal(loc = mean1, scale = sd, size = gray.shape)
    noise.append(x)
    
image_list = []

for i in range(10):
    image_list.append(gray + noise[i])
```

```python
# Display images with different gaussian noises
titles = ['Noise1', 'Noise2', 'Noise3', 'Noise4', 'Noise5', 'Noise6', 'Noise7', 'Noise8', 'Noise9', 'Noise10']
images = image_list

plt.figure(figsize=(20, 20))

for i in range(10):
    plt.subplot(4, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])

plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# Avg of Gaussian Images
avg_img = img_avg(image_list)

# Display image VS avg_img
titles = ['Grayscale', 'Noise Average']
images = [gray, avg_img]

plt.figure(figsize=(15, 15))

for i in range(2):
    plt.subplot(1, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])

plt.xticks([]),plt.yticks([])
plt.show()
```
