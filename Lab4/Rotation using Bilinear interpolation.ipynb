```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import math

# PISA Image
pisa_bgr = cv2.imread("PISA.jpg", 1)
# converting BGR to RGB
pisa = cv2.cvtColor(pisa_bgr, cv2.COLOR_BGR2RGB)
```

```python
def Biliear_interpolation_rotation(img,ang):  #anti-clockwise rotation.
    rad = (math.pi*ang)/180 #converting to radians
    w, h, l = img.shape
    xc = w//2
    yc = h//2
    res = np.zeros((w,h,l))    #height for rows and width for columns.
    print(res.shape)
    for k in range(0,l):
        for i in range(w):
            for j in range(h):
                    x = int((i - xc)*math.cos(rad) - (j - yc)*math.sin(rad))
                    y = int((i - xc)*math.sin(rad) + (j - yc)*math.cos(rad))
                    
                    xs = math.floor(x)
                    ys = math.floor(y)
                    a = x-xs
                    b = y-ys
                    idx_x = xs + xc
                    idx_y = ys + yc
                    if i+1<w and j+1<h and idx_x>=0 and idx_y>=0 and idx_x<w and idx_y<h:
                        res[idx_x][idx_y][k] = (1-a)*(1-b)*img[i][j][k] + (1-a)*b*img[i][j+1][k] + (1-b)*a*img[i+1][j][k] + a*b*img[i+1][j+1][k]
                        
    res = res.astype(np.int32)
    return res;
```

```python
ang = 8
img = pisa
res = Biliear_interpolation_rotation(img,ang)

plt.figure(figsize=(20,20))
plt.subplot(1,2,2)
plt.imshow(res)
plt.show()
```
