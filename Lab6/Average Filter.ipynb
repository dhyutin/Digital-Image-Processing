```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

lenna = cv2.imread("Lena.png") #reading lenna image
gray = cv2.cvtColor(lenna,cv2.COLOR_BGR2GRAY) #converting to grayscale

def noise(gray):  
    lst = []
    for x in range(1,11):
        gray2 = np.copy(gray)
        for i in range((gray.size//20)*x):
                y_coord=np.random.randint(0, gray.shape[0] - 1)
                x_coord=np.random.randint(0, gray.shape[1] - 1)
                 
                gray2[y_coord][x_coord] = 255
               
        for i in range((gray.size//20)*x):
                y_coord=np.random.randint(0, gray.shape[0] - 1)
                x_coord=np.random.randint(0, gray.shape[1] - 1)
                 
                gray2[y_coord][x_coord] = 0
        lst.append(gray2)     
        
    return lst
```

```python
img = noise(gray)

#printing the images
images = img
titles = ["noise img1","noise img2","noise img3","noise img4","noise img5","noise img6","noise img7","noise img8","noise img9","noise img10"]

plt.figure(figsize=(20,20))

for i in range(10):
    plt.subplot(5,2,i+1)
    plt.imshow(images[i],cmap="gray")
    plt.title(titles[i])
plt.show()
```

```python
filters = []
size = [3,5,7]
for i in size:
    m = np.ones([i,i], dtype = int)
    m = m * (1/i**2)
    filters.append(m)

final = []
for i in img:
    for j in filters:
        res = cv2.filter2D(i,-1,j)
        final.append(res)
```

```python
#printing the images
images = final
titles = ['img1 -3','img1 -5','img1 -7',
         'img2 -3','img2 -5','img2 -7',
         'img3 -3','img3 -5','img3 -7',
         'img4 -3','img4 -5','img4 -7',
         'img5 -3','img5 -5','img5 -7',
         'img6 -3','img6 -5','img6 -7',
         'img7 -3','img7 -5','img7 -7',
         'img8 -3','img8 -5','img8 -7',
         'img9 -3','img9 -5','img9 -7',
         'img10 -3','img10 -5','img10 -7']
plt.figure(figsize=(15,40))

for i in range(len(final)):
    plt.subplot(10,3,i+1)
    plt.imshow(images[i],cmap="gray")
    plt.title(titles[i])
plt.show()
```
